```python
### 가로 세로 크기 1/3로 줄이기
for i in range(0, 5000) : 
    img = Image.open('./newdata/raw/train/images/image_{:04d}.png'.format(i+1))
    img = img.resize((int(img.width/3), int(img.height/3)))
    img.save('./newdata/raw/train/images/image_{:04d}.png'.format(i+1))
```

```python
width_sum = 0
height_sum = 0

for i in range(0, 5000) : 
    img = Image.open('./newdata/raw/train/images/image_{:04d}.png'.format(i+1))
    width_sum += img.width
    height_sum += img.height

print("before | aihub train 데이터 평균 width : {}".format(width_sum/5000*2))
print("before | aihub train 데이터 평균 height : {}".format(height_sum/5000*2))
print("after  | aihub train 데이터 평균 width : {}".format(width_sum/5000))
print("after  | aihub train 데이터 평균 height : {}".format(height_sum/5000))
```

```python
from PIL import Image
 
image1 = Image.open('image_01002.PNG')
image1.show()
 
image2 = Image.open('image_01003.PNG')
image2.show()


print(image1.size)
print(image2.size)
# image1 = image1.resize((450, 163))
# image2 = image2.resize((450, 163))
image1_size = image1.size
image2_size = image2.size

new_image = Image.new('RGB',(image1.size[0] + image2.size[0], max(image1.size[1],image2.size[1] )), (255,255,255))

new_image.paste(image1,(0,0))
new_image.paste(image2,(image1_size[0],0))
new_image.save("merged_image.jpg","JPEG")
new_image.show()
```
