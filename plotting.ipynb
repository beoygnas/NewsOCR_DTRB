```python
f = open('./1.txt', 'r')

time = []
train_loss = []
valid_loss = []

while True:
    line = f.readline()
    if not line: break
    else: 
        if line[0] == '[' :
            train_loss.append(float(line.split(':')[1][0:8]))
            valid_loss.append(float(line.split(':')[2][0:8]))
            time.append(int(line.split(':')[3].split('.')[0][1:]))
f.close()
```

```python
import matplotlib.pyplot as plt

plt.plot(time, valid_loss, 'b',label='valid_loss')
plt.plot(time, train_loss, 'r', label= 'train_loss')
plt.xlabel('time')
plt.ylabel('loss')
plt.legend()
plt.savefig('None-VGG-BiLSTM-CTC.png')
```

```python
f = open('./log_demo_result.txt', 'r')

headline = []
column = [[], []] 
sentence = []
word_idx = 1

while True:
    line = f.readline()
    if not line: break
    else: 
        if 'column' in line : 
            col_idx = int(line.split('\t')[0].split('/')[-1].split('.')[0].split('_')[0][1:])
            st_idx = int(line.split('\t')[0].split('/')[-1].split('.')[0].split('_')[1][1:])
            end = line.split('\t')[0].split('/')[-1].split('.')[0].split('_')[2][-1]
            word_idx = 0
            
            if end == 'e' : 
                word_idx = int(line.split('\t')[0].split('/')[-1].split('.')[0].split('_')[2][:-1])
            else : 
                word_idx = int(line.split('\t')[0].split('/')[-1].split('.')[0].split('_')[2]) 

            word = line.split('\t')[1]
            
            if word_idx == 1 or end == 'e':
                if end == 'e' : sentence.append(word.replace(" ", ""))
                column[col_idx].append(sentence)
#                 print(sentence)
                sentence = []
            
            if end != 'e':
                sentence.append(word.replace(" ", ""))
            

            
            
        elif 'headline' in line : 
            st_idx = int(line.split('\t')[0].split('/')[-1].split('.')[0].split('_')[0][1:])
            end = line.split('\t')[0].split('/')[-1].split('.')[0].split('_')[1][-1]
            word_idx = 0
            
            if end == 'e' : 
                word_idx = int(line.split('\t')[0].split('/')[-1].split('.')[0].split('_')[1][:-1])
            else : word_idx = int(line.split('\t')[0].split('/')[-1].split('.')[0].split('_')[1])

            word = line.split('\t')[1]
#             print(word)
            if word_idx == 1 or end == 'e':
                if end == 'e' : sentence.append(word.replace(" ", ""))
                headline.append(sentence)
#                 print(sentence)
                sentence = []
            
            if end != 'e':
                sentence.append(word.replace(" ", ""))
            
f.close()
        
f = open('./results.txt', 'w')
        
for head in headline : 
    if head == [] : f.write("HEADLINE\n")
    else : 
        for word in head :
            f.write("{} ".format(word))

tmp_idx = 1
for col in column : 
    f.write("\n\nCOLUMN{}\n".format(tmp_idx))
    tmp_idx += 1
    for st in col :
        print(st)
        for word in st :
            f.write("{} ".format(word))
        
f.close()

print(column[0])
            
# for column 
            
```
